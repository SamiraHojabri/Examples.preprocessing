```python
smartphones.rename(index=smartphones.Name, inplace=True)
smartphones.drop(['Name','Company'], axis=1, inplace=True)
smartphones
```

```python
smartphones_new = pd.get_dummies(smartphones)#This method turns the Non-nemeric features into numeric features
smartphones_new
```

```python
smartphones_new.drop('OS_windows', axis=1, inplace=True)#the OS_windows column has a lot of Nan-value,so we drop it
smartphones_new
```

```python
from sklearn.preprocessing import scale,normalize,minmax_scale
```

```python
scale_data = scale(smartphones_new)#output is not in form of a dataframe like before,it's in array form
scale_data
```

```python
scale_data = scale(smartphones_new)#standardizing data with "scale"method
df_data = pd.DataFrame(scale_data, index=smartphones_new.index, columns=smartphones_new.columns)
#by using this code we can have a form of a dataframe as an output of "scale" method 
df_data
```

```python
normal_data = normalize(smartphones_new, norm='l2', axis=0)#standardizing data with "normalize"method
df_data1 = pd.DataFrame(normal_data, index = smartphones_new.index, columns = smartphones_new.columns)
df_data1
```

```python
minmax_scale = minmax_scale(smartphones_new, feature_range=(0,1))#standardizing data with "minmax_scale"method
df_data2 = pd.DataFrame(minmax_scale, index=smartphones_new.index, columns=smartphones_new.columns)
df_data2
```
